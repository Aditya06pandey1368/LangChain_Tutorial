## RunnableLambda

`RunnableLambda` is used to add custom Python functions into a LangChain workflow.

Example:

```python
from langchain_core.runnables import RunnableLambda

def to_upper(text):
    return text.upper()

uppercase = RunnableLambda(to_upper)
```

Flow:

```text
Prompt
   ↓
Model
   ↓
Parser
   ↓
RunnableLambda
```

### Use Cases

- Text Processing
- Data Transformation
- Calculations
- Custom Logic

---

## RunnablePassthrough

`RunnablePassthrough` passes the input forward without modifying it.

Example:

```python
from langchain_core.runnables import RunnablePassthrough

RunnablePassthrough().invoke("Hello")
```

Output:

```text
Hello
```

### Why Use It?

Used when we want to preserve the original input while performing additional operations in a chain.

---

````python
from langchain_core.runnables import RunnableLambda
from langchain_groq import ChatGroq
from groq.types import model
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser 

parser = StrOutputParser()

model = ChatGroq(
    model="llama-3.3-70b-versatile"
)

prompt = ChatPromptTemplate.from_template(
    "Explain {topic} in short"
)

#Function to count letters
def count(text):
    return len(text.split())


counter = RunnableLambda(count)

chain = prompt | model | parser | counter

response = chain.invoke({
    "topic" : "Runnable Lambda"
})
print(response)
````
