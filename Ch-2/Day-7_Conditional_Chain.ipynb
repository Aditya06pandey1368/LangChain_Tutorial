## RunnableBranch (Conditional Chain)

`RunnableBranch` is used to execute **only one chain** based on a condition.

It works like Python's `if-elif-else`.

---

## Why Use RunnableBranch?

- Route requests to different chains
- Execute only the required chain
- Useful for conditional workflows

Example:

- Programming Question → Programming Chain
- Math Question → Math Chain
- Others → General Chain

---

## Basic Syntax

```python
branch = RunnableBranch(
    (condition1, chain1),
    (condition2, chain2),
    default_chain
)
```

Invoke:

```python
response = branch.invoke(
    {"question": "What is Java?"}
)
```

---

## Flow

```text
             Input
               │
        Check Condition
               │
      ┌────────┼────────┐
      ▼        ▼        ▼
  Chain 1   Chain 2   Default
```

---

## Lambda Function

```python
lambda x: "java" in x["question"].lower()
```

Equivalent to:

```python
def check(x):
    return "java" in x["question"].lower()
```

- `lambda` → Fixed Python keyword.
- `x` → Variable name (can be changed).

---

## Important

- Conditions are checked **from top to bottom**.
- The **first matching condition** is executed.
- If no condition matches, the **default chain** runs.

---


````python
#Imports
from langchain_groq import ChatGroq
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnableBranch
````

````python
model = ChatGroq(
    model="llama-3.3-70b-versatile"
)

parser = StrOutputParser()
````

````python
programming_prompt = ChatPromptTemplate.from_template(
    """
    You are an expert programming teacher.

    Answer the following question:

    {question}
    """
)

math_prompt = ChatPromptTemplate.from_template(
    """
    You are an expert mathematics teacher.

    Answer the following question:

    {question}
    """
)

general_prompt = ChatPromptTemplate.from_template(
    """
    You are a helpful AI assistant.

    Answer the following question:

    {question}
    """
)
````

````python
programming_chain = programming_prompt | model | parser

math_chain = math_prompt | model | parser

general_chain = general_prompt | model | parser
````

````python
#Runnable Branch

branch = RunnableBranch(
    (
        lambda x : "programming" in x["question"].lower() or "java" in x["question"].lower() , 
        programming_chain
    ),
    (
        lambda x : "math" in x["question"].lower() or "integration" in x["question"].lower(),
        math_chain 
    ),
    general_chain
)
````

````python
response = branch.invoke(
    {
        "question" : "What is java"
    }
)

print(response)
````
