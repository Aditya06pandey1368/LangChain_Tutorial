## Structured Outputs

Structured Output means forcing the LLM to return data in a predefined structure instead of normal text.

Example:

Instead of:

```text
Aditya Pandey knows Java and has a CGPA of 7.42.
```

We get:

```python
Student(
    name="Aditya Pandey",
    cgpa=7.42,
    skills=["Java", "React", "MongoDB"]
)
```

---

## Why Use Structured Outputs?

- Consistent responses
- Easy to access data
- Useful for APIs
- Useful for Databases
- Useful for Agents and LangGraph

---

## What is Pydantic?

Pydantic is a Python library used to define and validate the structure of data.

It helps ensure that the output follows a fixed schema.

Example:

```python
class Person(BaseModel):
    name: str
    age: int
```

---

## What is BaseModel?

`BaseModel` is the parent class of all Pydantic models.

It is used to create schemas (blueprints) for structured data.

Example:

```python
class Student(BaseModel):
    name: str
    cgpa: float
```

---

## What is Field()?

`Field()` provides additional information about a field.

Example:

```python
name: str = Field(
    description="Student's Name"
)
```

It helps the LLM understand the purpose of the field more clearly.

---

## Structured Output Flow

```text
User Input
      ↓
LLM
      ↓
Pydantic Schema
      ↓
Structured Output
```

---

## Important Difference

### Normal LLM

```python
response = model.invoke(...)
```

Returns:

```python
AIMessage
```

Access using:

```python
response.content
```

---

### Structured Output

```python
response = structured_model.invoke(...)
```

Returns:

```python
Student Object
```

Access using:

```python
response.name
response.cgpa
response.skills
```

No `.content` is required.

---



````python
from dotenv import load_dotenv
from langchain_groq import ChatGroq
from pydantic import BaseModel, Field
from typing import List

load_dotenv()

class Student(BaseModel):
    name: str = Field(
        description="Student's Name"
    )
    cgpa: float = Field(
        description="Student's CGPA"
    )
    skills: List[str] = Field(
        description="List of technical skills"
    )

model = ChatGroq(
    model="llama-3.3-70b-versatile"
)

structured_model = model.with_structured_output(Student)

response = structured_model.invoke(
    """
    Aditya Pandey is a final year B.Tech CSE student.
    He knows Java, React, Node.js and MongoDB.
    His CGPA is 7.42.
    """
)

print(response)

print(response.name)
print(response.cgpa)
print(response.skills)

print(type(response))
print(type(response.skills))
````
