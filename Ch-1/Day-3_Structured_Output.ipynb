## 1. Load Environment Variables

```python
from dotenv import load_dotenv

load_dotenv()
```

Loads the API key from the `.env` file into the environment.

---

## 2. Import Required Libraries

```python
from langchain_groq import ChatGroq
from pydantic import BaseModel, Field
from typing import List
```

- **ChatGroq** → LLM provided by Groq.
- **BaseModel** → Creates a schema for structured data.
- **Field** → Adds metadata (description) to each field.
- **List** → Specifies that a field contains multiple values.

---

## 3. Create a Pydantic Schema

```python
class Student(BaseModel):
```

Creates a schema named **Student**.

Every structured response from the LLM should follow this schema.

---

### name

```python
name: str = Field(
    description="Student's Name"
)
```

- `str` → Name must be a string.
- `description` helps the LLM understand what to extract.

---

### cgpa

```python
cgpa: float = Field(
    description="Student's CGPA"
)
```

- `float` → Decimal number.

Example:

```python
7.42
```

---

### skills

```python
skills: List[str] = Field(
    description="List of technical skills"
)
```

`List[str]` means:

```python
[
    "Java",
    "React",
    "MongoDB"
]
```

---

## 4. Create the Model

```python
model = ChatGroq(
    model="llama-3.3-70b-versatile"
)
```

Creates the LLM.

---

## 5. Convert Normal Model into Structured Model

```python
structured_model = model.with_structured_output(Student)
```

Instead of returning normal text, the model now returns a **Student object** following the schema.

---

## 6. Invoke the Model

```python
response = structured_model.invoke(
    """
    Aditya Pandey is a final year B.Tech CSE student.
    He knows Java, React, Node.js and MongoDB.
    His CGPA is 7.42.
    """
)
```

The LLM extracts the information and fills the schema automatically.

---

## Important Difference

### Normal LLM

```python
response = model.invoke(...)
```

Returns:

```python
AIMessage
```

Access using:

```python
response.content
```

---

### Structured Output

```python
response = structured_model.invoke(...)
```

Returns:

```python
Student Object
```

Access using:

```python
response.name
response.cgpa
response.skills
```

No `.content` is required.

````python
from dotenv import load_dotenv
from langchain_groq import ChatGroq
from pydantic import BaseModel, Field
from typing import List

load_dotenv()

class Student(BaseModel):
    name: str = Field(
        description="Student's Name"
    )
    cgpa: float = Field(
        description="Student's CGPA"
    )
    skills: List[str] = Field(
        description="List of technical skills"
    )

model = ChatGroq(
    model="llama-3.3-70b-versatile"
)

structured_model = model.with_structured_output(Student)

response = structured_model.invoke(
    """
    Aditya Pandey is a final year B.Tech CSE student.
    He knows Java, React, Node.js and MongoDB.
    His CGPA is 7.42.
    """
)

print(response)

print(response.name)
print(response.cgpa)
print(response.skills)

print(type(response))
print(type(response.skills))
````
