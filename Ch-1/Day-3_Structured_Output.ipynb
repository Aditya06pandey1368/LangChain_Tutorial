# 🚀 Day-3 Code Explanation

## 1. Load Environment Variables

```python
load_dotenv()
```

Loads the API key from the `.env` file.

---

## 2. Import Libraries

```python
ChatGroq
BaseModel
Field
List
```

- **ChatGroq** → LLM
- **BaseModel** → Creates a schema
- **Field** → Adds description to fields
- **List** → Represents a list of values

---

## 3. Create a Schema

```python
class Student(BaseModel):
```

Defines the structure of the output.

```python
name: str
cgpa: float
skills: List[str]
```

---

## 4. Structured Model

```python
structured_model = model.with_structured_output(Student)
```

Converts the normal LLM into one that returns a **Student object**.

---

## 5. Invoke

```python
response = structured_model.invoke(...)
```

The LLM extracts the information according to the schema.

---

## 6. Access Data

```python
print(response.name)
print(response.cgpa)
print(response.skills)
```

No `.content` is required.

---

## 7. Check Types

```python
print(type(response))
```

```python

```

```python
print(type(response.skills))
```

Output:


---

## Important

Normal Model:

```python
response = model.invoke(...)
```

Returns:

```python
AIMessage
```

Access:

```python
response.content
```

---

Structured Model:

```python
response = structured_model.invoke(...)
```

Returns:

```python
Student
```

Access:

```python
response.name
response.cgpa
response.skills
```

````python
from dotenv import load_dotenv
from langchain_groq import ChatGroq
from pydantic import BaseModel, Field
from typing import List

load_dotenv()

class Student(BaseModel):
    name: str = Field(
        description="Student's Name"
    )
    cgpa: float = Field(
        description="Student's CGPA"
    )
    skills: List[str] = Field(
        description="List of technical skills"
    )

model = ChatGroq(
    model="llama-3.3-70b-versatile"
)

structured_model = model.with_structured_output(Student)

response = structured_model.invoke(
    """
    Aditya Pandey is a final year B.Tech CSE student.
    He knows Java, React, Node.js and MongoDB.
    His CGPA is 7.42.
    """
)

print(response)

print(response.name)
print(response.cgpa)
print(response.skills)

print(type(response))
print(type(response.skills))
````
