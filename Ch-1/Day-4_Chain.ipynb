## Chains in LangChain

A Chain is used to connect multiple LangChain components together.

Instead of manually passing the output of one component to another, LangChain does it automatically.

---

## Why Use Chains?

Without Chain:

```python
formatted_prompt = prompt.invoke(...)
response = model.invoke(formatted_prompt)
```

With Chain:

```python
chain = prompt | model
response = chain.invoke(...)
```

Cleaner and easier to manage.

---

## Chain Flow

```text
Prompt
   ↓
Model
   ↓
Response
```

---

## The `|` Operator

The `|` operator connects components together.

```python
chain = prompt | model
```

Meaning:

```text
Output of Prompt
        ↓
Input of Model
```

---

## Output Parsers

Parsers can be added to a chain.

```python
chain = prompt | model | parser
```

Flow:

```text
Prompt
   ↓
Model
   ↓
Parser
   ↓
Final Output
```

---

## StrOutputParser

Converts the model output into a normal string.

```python
from langchain_core.output_parsers import StrOutputParser

parser = StrOutputParser()
```

Without Parser:

```python
response.content
```

With Parser:

```python
response
```

returns a normal string.

---
---


### Easiest Chain

````python
from langchain_groq import ChatGroq
from groq.types import model
from langchain_core.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate.from_template(
    "Explain {topic} in simple language."
)
model = ChatGroq(
    model="llama-3.3-70b-versatile"
)
chain = prompt | model

response = chain.invoke(
    {
        "topic" : "Chain in Langchain"
    }
)

#print(response.content)
print(type(chain))
````

# Chain with a Parser

## Output Parsers

Output Parsers are used to transform the output of an LLM into a desired format.

---

### StrOutputParser

Converts the model output into a normal Python string.

```python
from langchain_core.output_parsers import StrOutputParser

parser = StrOutputParser()
```

---

### Chain with Parser

```python
chain = prompt | model | parser
```

Flow:

```text
Prompt
   ↓
Model
   ↓
Parser
   ↓
String Output
```

---

### Important

Without Parser:

```python
response = chain.invoke(...)
```

Returns:

```python
AIMessage
```

Access using:

```python
response.content
```

With `StrOutputParser`:

```python
response = chain.invoke(...)
```

Returns:

```python
str
```

Access directly:

```python
print(response)
```

---


````python
from langchain_core.output_parsers import StrOutputParser 

parser = StrOutputParser()

chain = prompt | model | parser

response = chain.invoke({
    "topic" : "What is parser in langchain"
})

print(response)
````
