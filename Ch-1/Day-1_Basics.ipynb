## What is LangChain?

LangChain is a framework used to build applications powered by Large Language Models (LLMs) such as GPT, Gemini, Claude, and Llama.

## Why are we using LangChain?

LangChain provides useful components such as:

- Prompt Templates
- Chains
- Memory
- Agents
- RAG (Retrieval-Augmented Generation)
- Tool Calling

It makes AI applications easier to build and manage.

## Prompt Templates

Prompt Templates allow dynamic prompts using placeholders.

Example:

```python
{question}

````python
from dotenv import load_dotenv
from langchain_groq import ChatGroq
from langchain_core.prompts import ChatPromptTemplate

load_dotenv()

prompt = ChatPromptTemplate.from_template(
    "Your are a funny teacher who teaches everything in a funny way and the questions are {question}"
)

formatted_prompt = prompt.invoke({"question":"What is langchain"})

model = ChatGroq(
    model="llama-3.3-70b-versatile"
)
response = model.invoke(formatted_prompt)

print(response.content)
````
