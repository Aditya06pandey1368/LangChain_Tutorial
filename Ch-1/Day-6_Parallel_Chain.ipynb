## RunnableParallel

`RunnableParallel` is used to execute multiple independent chains simultaneously using the same input.

It returns the output of all chains as a **dictionary**.

---

## Why Use RunnableParallel?

- Faster execution
- Multiple tasks at the same time
- Useful when tasks are independent

Example:

- Explain Topic
- Generate Notes
- Create Quiz
- Interview Questions

All run in parallel.

---

## Basic Syntax

```python
parallel_chain = RunnableParallel(
    explanation=explanation_chain,
    notes=notes_chain
)
```

Invoke:

```python
response = parallel_chain.invoke(
    {"topic": "LangChain"}
)
```

Output:

```python
{
    "explanation": "...",
    "notes": "..."
}
```

---

## Flow

```text
            Input
              │
      ┌───────┴───────┐
      ▼               ▼
Chain 1          Chain 2
      ▼               ▼
      └───────┬───────┘
              ▼
      Dictionary Output
```

---

## Important

- Every branch receives the **same input**.
- The input type depends on what the branches expect (`str`, `dict`, `int`, etc.).
- `RunnableParallel` itself does **not** require a dictionary; it simply forwards the same input to all branches.

---



````python
#Imports
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser 
from langchain_core.runnables import RunnableParallel
from langchain_groq import ChatGroq

model = ChatGroq(
    model="llama-3.3-70b-versatile"
)

parser = StrOutputParser()
````

````python
#Explanation prompt
explanation_prompt = ChatPromptTemplate.from_template(
    "Explain {topic} in simple"
)

#Notes
notes_prompt = ChatPromptTemplate.from_template(
    "Create short notes on {topic}"
)

#Interview Question
interview_prompt = ChatPromptTemplate.from_template(
    "Generate 5 interview question on {topic}"
)

#Application
application_prompt = ChatPromptTemplate.from_template(
    "Give 5 real-world application of {topic}"
)
````

````python
#Chains
explanation_chain = explanation_prompt | model | parser

notes_chain = notes_prompt | model | parser

interview_chain = interview_prompt | model | parser

application_chain = application_prompt | model | parser
````

````python
parallel_chain = RunnableParallel(
    explanation=explanation_chain,
    notes=notes_chain,
    interview_questions=interview_chain,
    applications=application_chain
)
````

````python
response = parallel_chain.invoke(
    {
        "topic": "LangChain"
    }
)

print(response["applications"])
````
