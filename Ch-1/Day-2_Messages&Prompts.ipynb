## Messages in LangChain

Messages are used to structure conversations between the user and the AI.

Unlike Prompt Templates, messages separate instructions, user input, and AI responses.

---

## Types of Messages

### SystemMessage

Defines the behavior of the AI.

Example:

```python
SystemMessage(
    content="You are a funny teacher."
)
```

---

### HumanMessage

Represents the user's input.

Example:

```python
HumanMessage(
    content="What is LangChain?"
)
```

---

### AIMessage

Represents the AI's response.

Example:

```python
AIMessage(
    content="LangChain is a framework for building LLM applications."
)
```

---

## Why Use Messages?

Messages create structured conversations.

```text
System → Instructions
Human → User Input
AI → Model Response
```

They are mainly used in:

- Chatbots
- Memory
- Agents
- LangGraph

---

## Prompt Template vs Messages

### Prompt Template

Creates a formatted prompt using placeholders.

```python
prompt = ChatPromptTemplate.from_template(
    "Explain {topic}"
)
```

### When do we use Prompt Templates?

- When we need variables/placeholders.
- When we want to generate prompts dynamically.
- Information is generally sent as a single formatted text (paragraph) to the LLM.

Example:

```text
You are a funny teacher.
Explain LangChain.
```

---

### Messages

Create structured conversations.

```python
[
    SystemMessage(...),
    HumanMessage(...)
]
```

### When do we use Messages?

- When we need conversations.
- When we need chat history.
- When building chatbots, agents, memory, and LangGraph applications.

Example:

```text
System: You are a funny teacher.

Human: What is LangChain?

AI: LangChain is a framework...
```

---

## Important

Prompt Templates and Messages are not competitors.

They are often used together.

---

## Current Flow

```text
SystemMessage
       ↓
HumanMessage
       ↓
LLM
       ↓
AIMessage
```

---



````python
from dotenv import load_dotenv
from langchain_groq import ChatGroq
from langchain_core.messages import SystemMessage, HumanMessage

load_dotenv()

model = ChatGroq(
    model="llama-3.3-70b-versatile"
)

messages = [
    SystemMessage(
        content="You are a funny teacher who teaches everything in a funny way."
    ),
    HumanMessage(
        content="What is LangChain?"
    )
]

response = model.invoke(messages)

print(response.content)
````

## Prompt Template Version (Comparison)

````python
from dotenv import load_dotenv
from langchain_groq import ChatGroq
from langchain_core.prompts import ChatPromptTemplate

load_dotenv()

prompt = ChatPromptTemplate.from_template(
    """
    You are a funny teacher.

    Answer the following question in a humorous way:

    {question}
    """
)

model = ChatGroq(
    model="llama-3.3-70b-versatile"
)

chain = prompt | model

response = chain.invoke(
    {
        "question": "What is LangChain?"
    }
)

print(response.content)
````
